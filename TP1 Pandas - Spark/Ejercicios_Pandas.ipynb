# Ejercicio 35

Para los usuarios nativos (o superior) en español obtenga una serie cuyo índice sea cada uno de los otros idiomas que sabe y valor sea el nivel promedio (tomando N=4.5) (⭐⭐)

```python
lenguajes = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/languages.csv")
#Primero chequeo q no haya filas duplicadas y veo que no hay (omito operacion)
#Paso a numero y reemplazo n por el valor aclarado
def replace_babel_level(x):
    if (x == "N"):
      return 4.5
    return float(x)

#Borro filas que no tienen aclarado el lenguaje
lenguajes = lenguajes.query("babel_lang.notnull()")
lenguajes.loc[:,'babel_level'] = [replace_babel_level(x) for x in lenguajes.babel_level]
lenguajes
```

```python
#Busco que usuarios hablan nativo o mejor español
usuarios = lenguajes.query("babel_level >= 4.5 and babel_lang == 'es' ")
usuarios = usuarios['babel_user']
usuarios
```

```python
#Me quedo solo con los usuarios que esten en esa lista, y saco las filas de español
lenguajes = lenguajes[lenguajes.babel_user.isin(usuarios)]
lenguajes = lenguajes.query("babel_lang != 'es'")
lenguajes = lenguajes[['babel_lang','babel_level']]
lenguajes
```

```python
#Resultado
promedio = lenguajes.groupby(['babel_lang']).mean()
promedio  = promedio['babel_level']
promedio
```

# Ejercicio 46

46. Si decimos que la ubicación de una categoría es el promedio de la latitud y longitud de sus contenidos geolocalizados que son miembros de ella (si es que tiene): ¿Cuales son las dos categorías más cercanas?

```python

import functools as functools
```

```python
geo_tags = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/geo_tags.csv")
categorias = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/categorylinks.csv")
```

```python
#Me quedo con las columnas que me interesan
geo_tags = geo_tags.loc[:,['gt_page_id','gt_lat','gt_lon']]

categorias = categorias.loc[:,['cl_from','cl_to']]

#Renombro columnas que significan lo mismo para menor confusion
geo_tags.rename(columns = {'gt_page_id':'id_contenido'}, inplace = True)
categorias.rename(columns = {'cl_from':'id_contenido'}, inplace = True)
categorias.rename(columns = {'cl_to':'categoria'}, inplace = True)
```

Ubicacion de una categoria = promedio de latitud y longitud de sus contenidos.

```python
#Me quedo con las categorias que tienen contenidos y los contenidos que tienen categorias
datos_categorias = geo_tags.merge(categorias, how = 'inner', on='id_contenido')
```

```python
#Agrupo las latitudes y longitudes de cada categoria sacando su promedio
datos_categorias = datos_categorias.groupby(['categoria'], as_index= False).agg({'gt_lat': 'mean', 'gt_lon': 'mean'})

#guardo la ubicacion como una tupla y agrupo por ubicacion
datos_categorias['ubicacion']= list(zip(datos_categorias.gt_lat,datos_categorias.gt_lon))
datos_categorias = datos_categorias[['categoria', 'ubicacion']]
datos_categorias
```

```python
#Junto aquellas categorias q comparten ubicacion y me quedo con aquellas ubicaciones con mas de una categoria
datos_categorias = datos_categorias.groupby('ubicacion').agg({'categoria':lambda x: list(x)})
datos_categorias = datos_categorias[datos_categorias.categoria.map(len)>1]

datos_categorias
```

Las dos categorias mas cercanas son cualquier dos categorias pertenecientes a una misma ubicacion.
